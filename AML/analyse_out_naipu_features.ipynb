```python
import pandas as pd
import numpy as np
```

```python
base_path = "NIAPU/NIAPU/"
```

```python
seed_features = pd.read_csv(base_path + "seed_features", header=None, sep="\t")
seed_features
```

```python
out_genes = pd.read_csv(base_path + "src/out_genes", header=None, sep=" ")
out_genes
```

```python
positive_seeds = out_genes[out_genes.iloc[:, 2] > 0]
positive_seeds.to_csv(base_path + "naipu_preprocessed_positive_seeds.tsv", sep="\t", index=None)
positive_seeds
```

```python
negative_seeds = out_genes[out_genes.iloc[:, 2] == 0]
negative_seeds.to_csv(base_path + "naipu_preprocessed_negative_seeds.tsv", sep="\t", index=None)
negative_seeds
```
