```python
import pandas as pd
import numpy as np
## https://academic.oup.com/bioinformatics/article/39/2/btac848/7023926?login=false
# https://www.disgenet.org/browser/0/0/0/0/geneid__7124-source__CURATED/_b./
# https://www.disgenet.org/browser/0/1/0/C0026998/
```

```python
base_path = "../nanodiag_datasets/GSE175758/"
```

```python
non_rand_cpgs = pd.read_csv(base_path + "non-randomly-demethylated-CpGs.tsv", sep="\t")
non_rand_cpgs
```

```python
non_rand_cpgs.columns = [item.strip() for item in non_rand_cpgs.columns]
non_rand_cpgs.columns
```

```python
cpgs_associaton_scores = non_rand_cpgs[["CpG", "delta beta"]]
#cpgs_associaton_scores["association_score"] = np.abs(non_rand_cpgs["delta beta"])
#cpgs_associaton_scores = cpgs_associaton_scores.drop(columns=["delta beta"])
cpgs_associaton_scores.columns = ["feature_name", "association_score"]
cpgs_associaton_scores["association_score"] = np.abs(cpgs_associaton_scores["association_score"])
cpgs_associaton_scores
```

```python
genes_symbols_diff_exp_meth = pd.read_csv(base_path + "genes_symbol_methylation_expression.tsv", sep="\t")
genes_symbols_diff_exp_meth
```

```python
genes_association_scores = genes_symbols_diff_exp_meth[["Gene symbol", "logFoldChange"]]
#genes_association_scores["association_score"] =  genes_symbols_diff_exp_meth["logFoldChange"] #columns=
#genes_association_scores = genes_association_scores.drop(columns=["logFoldChange"])
genes_association_scores.columns = ["feature_name", "association_score"]
genes_association_scores
```

```python
combine_cpg_genes = pd.concat([cpgs_associaton_scores, genes_association_scores], axis=0)
combine_cpg_genes
```

```python
final_feature_names = pd.read_csv(base_path + "final_feature_names.tsv", sep="\t")
final_feature_names
```

```python
seed_feature_names = list()
seed_feature_association_scores = list()

for i, row in enumerate(final_feature_names.iterrows()):
    name = row[1]["feature_names"].split("_")
    #print(name)
    if name[0] in combine_cpg_genes["feature_name"].tolist() and \
        name[1] in combine_cpg_genes["feature_name"].tolist():
        seed_feature_names.append(row[1]["feature_names"])
        score = combine_cpg_genes[combine_cpg_genes["feature_name"] == name[0]]["association_score"].iloc[0]
        score += combine_cpg_genes[combine_cpg_genes["feature_name"] == name[1]]["association_score"].iloc[0]
        seed_feature_association_scores.append(score)
        #print("CpG+Gene:", print(name[0], name[1]), score)
    elif name[0] in combine_cpg_genes["feature_name"].tolist():
        seed_feature_names.append(row[1]["feature_names"])
        score = combine_cpg_genes[combine_cpg_genes["feature_name"] == name[0]] #["association_score"]
        seed_feature_association_scores.append(score["association_score"].iloc[0])
        #print("CpG:", print(name[0]), score)
        #print()
        #print(score["association_score"].iloc[0])
        #break
    elif name[1] in combine_cpg_genes["feature_name"].tolist():
        seed_feature_names.append(row[1]["feature_names"])
        score = combine_cpg_genes[combine_cpg_genes["feature_name"] == name[1]] #["association_score"]
        seed_feature_association_scores.append(score["association_score"].iloc[0])
        #print("Gene:")
        #print("Name:", name[1])
        #print(score[2])
        #break
```

```python
df_seed_features = pd.DataFrame(zip(seed_feature_names, seed_feature_association_scores), columns=["feature_name", "score"])
df_seed_features
```
