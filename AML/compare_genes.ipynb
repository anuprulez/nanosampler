```python
import numpy as np
import pandas as pd
```

```python
base_path = "../nanodiag_datasets/GSE175758/"
```

```python
path_positive_seeds = "seed_features.tsv"
```

```python
df_pos_seeds = pd.read_csv(base_path + path_positive_seeds, sep="\t")
df_pos_seeds.head()
```

```python
feature_names = df_pos_seeds["feature_name"].tolist()
feature_names[:5]
```

```python
aml_gene_names = list(set([item.split("_")[1] for item in feature_names]))
aml_gene_names, len(aml_gene_names)
```

```python
df_aml_genes_m1_disgenet = pd.read_csv("AML_GDA_Disgenet/C0026998_AML_M1_disease_gda_summary.tsv", sep="\t")
df_aml_genes_m1_disgenet.head()
```

```python
df_aml_genes_m1_disgenet_gene_scores = df_aml_genes_m1_disgenet[["Gene", "Score_gda"]]
df_aml_genes_m1_disgenet_gene_scores.head()
```

```python
m1_genes = df_aml_genes_m1_disgenet_gene_scores["Gene"].tolist()
m1_genes[:5], len(m1_genes)
```

```python
df_aml_genes_m2_disgenet = pd.read_csv("AML_GDA_Disgenet/C1879321_AML_M2_disease_gda_summary.tsv", sep="\t")
df_aml_genes_m2_disgenet.head()
df_aml_genes_m2_disgenet_gene_scores = df_aml_genes_m2_disgenet[["Gene", "Score_gda"]]
df_aml_genes_m2_disgenet_gene_scores.head()
```

```python
m2_genes = df_aml_genes_m2_disgenet_gene_scores["Gene"].tolist()
m2_genes[:5], len(m2_genes)
```

```python
m2_genes.extend(m1_genes)
m1_m2_genes = m2_genes
m1_m2_genes = list(set(m1_m2_genes))
m1_m2_genes[:5], len(m1_m2_genes)
```

```python
intersection = list(set(m1_m2_genes).intersection(set(aml_gene_names)))
intersection, len(intersection)
```
