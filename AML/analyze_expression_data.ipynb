```python
import pandas as pd
import numpy as np
```

```python
base_path = "../nanodiag_datasets/"
df_expression_data = pd.read_csv(base_path + "df_expression_data_transpose.csv", sep="\t")
df_expression_data
```

```python
sample_names = df_expression_data["Unnamed: 0"]
labels = df_expression_data["treatment_label"]
sample_names, labels
```

```python
df_expression_data_feature = df_expression_data.drop(columns=["Unnamed: 0", "treatment_label"])
df_expression_data_feature
```

```python
from xgboost import XGBClassifier
from matplotlib import pyplot
```

```python
model = XGBClassifier()
model.fit(df_expression_data_feature, labels)
```

```python
# feature importance
df_expression_data_feature
# plot
pyplot.bar(range(len(model.feature_importances_)), model.feature_importances_)
pyplot.show()
```

```python
feature_imps = pd.DataFrame(zip(df_expression_data_feature.columns, model.feature_importances_), columns=["Gene", "Importance"])
```

```python
feature_imps
```

```python
feature_imps = feature_imps.sort_values(ascending=False, by="Importance")
feature_imps = feature_imps[feature_imps["Importance"] > 0.0]
feature_imps
```

```python
feature_imps.to_csv(base_path + "significant_genes.csv", sep="\t", index=None)
```

```python
groud_truth_genes = pd.read_csv(base_path + "ground_truth_genes.csv", sep="\t")
groud_truth_genes
```

```python
true_genes = groud_truth_genes["Gene symbol"].tolist()
```

```python
pred_genes = feature_imps["Gene"].tolist()
pred_genes = [item.strip() for item in pred_genes]
```

```python
intersection = list(set(true_genes).intersection(set(pred_genes)))
intersection
```

```python
import umap
import umap.plot
import matplotlib.pyplot as plt

# n_neighbors
# min_dist
# n_components
# metric

umap.plot.points(umap.UMAP(n_neighbors=50).fit(df_expression_data_feature), labels=labels)
```

```python
df_expression_data_feature_pred_genes = df_expression_data_feature[feature_imps["Gene"]]
umap.plot.points(umap.UMAP(n_neighbors=50).fit(df_expression_data_feature_pred_genes), labels=labels)
```
