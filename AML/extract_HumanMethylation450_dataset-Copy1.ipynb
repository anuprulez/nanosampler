```python
import pandas as pd
import numpy as np
```

```python
base_path = "../nanodiag_datasets/GSE175758/"
GSE175758_GEO_processed = base_path + "GSE175758_GEO_processed.txt"
```

```python
df_GSE175758_GEO_processed = pd.read_csv(GSE175758_GEO_processed, sep="\t")
```

```python
df_GSE175758_GEO_processed
```

```python
d8_features = list()
d0_features = list()

for col in df_GSE175758_GEO_processed.columns:
    if "blasts.d0" in col and "Detection.Pval" not in col:
        d0_features.append(col)
    if "blasts.d8" in col and "Detection.Pval" not in col:
        d8_features.append(col)

len(d0_features), len(d8_features)
```

```python
df_d0 = df_GSE175758_GEO_processed[d0_features]
df_d0
```

```python
df_d8 = df_GSE175758_GEO_processed[d8_features]
df_d8
```

```python
merged_do_d8 = pd.concat([df_d0, df_d8], axis=1)
merged_do_d8
```

```python
merged_do_d8_transpose = merged_do_d8.transpose()
merged_do_d8_transpose
```

```python
treatment_label = []
for i, row in merged_do_d8_transpose.iterrows():
    if "d0" in row.name:
        treatment_label.append(0)
    else:
        treatment_label.append(1)

#merged_do_d8_transpose["treatment_label"] = treatment_label
merged_do_d8_transpose.drop(columns=["treatment_label"])
```

```python
labels = merged_do_d8_transpose["treatment_label"]

features = merged_do_d8_transpose.drop(columns=["treatment_label"])
features
```

```python
import umap
import umap.plot

# n_neighbors
# min_dist
# n_components
# metric

mapper = umap.UMAP(n_neighbors=15).fit(features)
umap.plot.points(mapper, labels=labels)
```
