```python
import pandas as pd
import numpy as np
```

```python
base_path = "../nanodiag_datasets/"
sample_id_mapper_path = base_path + "GPL17586-45144-only-table.txt"
expression_data_path = base_path + "GSE171053_Greve_Andrieux_et_al_Expression_matrix.csv"
```

```python
df_mapper = pd.read_csv(sample_id_mapper_path, sep=",")
df_mapper
```

```python
df_expression_data = pd.read_csv(expression_data_path, sep=",")
df_expression_data
```

```python
mapper_ids = df_mapper["ID"]
reference_ids = mapper_ids.tolist()
len(reference_ids)
```

```python
expression_ids = df_expression_data["ID_REF"].tolist()
len(expression_ids)
```

```python
intersecting_ids = list(set(reference_ids).intersection(set(expression_ids)))
len(intersecting_ids)
```

```python
df_expression_data_filtered = df_expression_data.loc[df_expression_data["ID_REF"].isin(intersecting_ids)]
df_expression_data_filtered
```

```python
df_mapper_filtered = df_mapper.loc[df_mapper["ID"].isin(intersecting_ids)]
df_mapper_filtered
```

```python
gene_list = list()
for i, row in df_expression_data_filtered.iterrows():
    gene_name = df_mapper_filtered[df_mapper_filtered["ID"] == row["ID_REF"]]["gene_assignment"].values[0]
    gene_name = gene_name.split("//")[1]
    gene_list.append(gene_name)

    
```

```python
df_expression_data_filtered["genes"] = gene_list
df_expression_data_filtered
```

```python
df_expression_data_filtered = df_expression_data_filtered[df_expression_data_filtered["genes"] != " --- "]
df_expression_data_filtered
```

```python
df_expression_data_filtered_original = df_expression_data_filtered.transpose()
df_expression_data_filtered_original
```

```python
df_expression_data_filtered_original
```

```python
df_expression_data_filtered_original.reset_index(drop=True)
gene_names = df_expression_data_filtered_original["genes"]
df_expression_data_filtered_original_no_genes = df_expression_data_filtered_original.drop(columns=["genes"])
df_expression_data_filtered_original_no_genes
#df_expression_data_filtered.to_csv(base_path + "expression_data_filtered.csv", sep="\t")
```

```python
gene_names
```

```python
df_expression_data_transpose = df_expression_data_filtered_original_no_genes.transpose()
df_expression_data_transpose
```

```python
df_expression_data_transpose.columns = gene_names.tolist()
df_expression_data_transpose
```

```python
patients_list= df_expression_data_transpose.index.tolist()[:5]
patients_list
```

```python
treatment_label = []
for i, row in df_expression_data_transpose.iterrows():
    #print(row.name)
    if "d0" in row.name:
        treatment_label.append(0)
    else:
        treatment_label.append(1)

df_expression_data_transpose["treatment_label"] = treatment_label
df_expression_data_transpose
```
