```python
import pandas as pd
import numpy as np
```

```python
base_path = "../nanodiag_datasets/GSE175758/"
```

```python
df_merged_signals = pd.read_csv(base_path + "merged_signals.csv", sep="\t")
df_merged_signals
```

```python
labels = df_merged_signals["treatment_label"]
features = df_merged_signals.drop(columns=["treatment_label"])
```

```python
d0 = features[:25]
d0
```

```python
d8 = features[25:]
d8
```

```python
difference_d0_d8 = d0.reset_index().subtract(d8.reset_index(), axis = 1)  #d0.subtract(d8)
difference_d0_d8 = difference_d0_d8.drop(columns=["index"])
difference_d0_d8
```

```python
difference_d0_d8 = np.sqrt(np.square(difference_d0_d8))
difference_d0_d8
```

```python
std_deviation_per_feature = np.std(difference_d0_d8, axis=0)

std_deviation_per_feature.columns = ["Names", "STD"]
std_deviation_per_feature
```

```python
sorted_features = std_deviation_per_feature.sort_values(ascending=False)
sorted_features
```

```python
top_sorted_features = sorted_features[:10]
top_sorted_features
```

```python
last_sorted_features = sorted_features[::-1][:10]
last_sorted_features
```

```python
non_rand_cpgs = pd.read_csv(base_path + "non-randomly-demethylated-CpGs.tsv", sep="\t")
non_rand_cpgs
```

```python
all_cols = non_rand_cpgs.columns
all_cols = [item.strip() for item in all_cols]
non_rand_cpgs.columns = all_cols
non_rand_cpgs.columns
```

```python
genes_symbols_diff_exp_meth = pd.read_csv(base_path + "genes_symbol_methylation_expression.tsv", sep="\t")
genes_symbols_diff_exp_meth
```

```python
# find POS probe ids from original matrix using gene symbols and CpGs mentioned in https://www.nature.com/articles/s41375-023-01876-2
positive_probes_genes = list()
for index, col in enumerate(features.columns):
    cpg, gene = col.split("_")[0], col.split("_")[1]
    if gene in genes_symbols_diff_exp_meth["Gene symbol"].tolist() or cpg in non_rand_cpgs["CpG"].tolist():
        positive_probes_genes.append(col)
len(positive_probes_genes)
```

```python
positive_signals = features[positive_probes_genes]
positive_signals
```

```python
negative_cols = list(set(features.columns.tolist()).difference(positive_probes_genes))
print(len(features.columns.tolist()), len(negative_cols))
negative_signals = features[negative_cols]
negative_signals
```

```python
balanced_negative_signals = features[negative_signals.columns[:len(positive_probes_genes)]]
balanced_negative_signals
```

```python
negative_probes_genes = balanced_negative_signals.columns.tolist()
negative_probes_genes 
```

```python
#negative_probes_genes = negative_probes_genes.tolist()
df_neg = pd.DataFrame(negative_probes_genes, columns=["negative_probes_genes"])
df_neg.to_csv(base_path + "negative_probes_genes.tsv", index=None)
df_neg
```

```python
positive_probes_genes = positive_probes_genes
df_pos = pd.DataFrame(positive_probes_genes, columns=["positive_probes_genes"])
df_pos.to_csv(base_path + "positive_probes_genes.tsv", index=None)
df_pos
#positive_probes_genes.to_csv(base_path + "positive_probes_genes.tsv", sep="\t", index=None)
```

```python
probe_genes_mapping = features.columns.tolist()
probe_genes_mapping
```

```python
id_range = np.arange(0, len(probe_genes_mapping))
df_probe_genes_mapping = pd.DataFrame(zip(probe_genes_mapping, id_range), columns=["name", "id"])
df_probe_genes_mapping
```

```python
df_probe_genes_mapping.to_csv(base_path + "probe_genes_mapping_id.tsv", sep="\t", index=None)
```

```python
balanced_negative_signals.to_csv(base_path + "balanced_negative_signals.tsv", sep="\t", index=None)
positive_signals.to_csv(base_path + "positive_signals.tsv", sep="\t", index=None)
```

```python
combined_pos_neg_signals = pd.concat([positive_signals, balanced_negative_signals], axis=1)
combined_pos_neg_signals
```

```python
transposed_matrix = np.transpose(combined_pos_neg_signals)
transposed_matrix
```

```python
# Compute Pearson correlation matrix
correlation_matrix = np.corrcoef(transposed_matrix)
correlation_matrix, correlation_matrix.shape
```

```python
np.min(correlation_matrix)
```

```python
df_corr = pd.DataFrame(correlation_matrix)
df_corr.index = transposed_matrix.index
df_corr.columns = transposed_matrix.index
df_corr
```

```python
df_corr.to_csv(base_path + "df_pos_neg_probes_correlation_matrix.tsv", sep="\t")
```
