```python
probe_mapper = probe_mapper_full[["ID", "UCSC_RefGene_Name", "UCSC_RefGene_Group", "Forward_Sequence"]]
probe_mapper = probe_mapper.dropna()
probe_mapper
```

```python
#probe_mapper["Forward_Sequence_formatted"] = probe_mapper["Forward_Sequence"].replace("[", "").replace("]", "")
import re
# Define the regular expression pattern
pattern = r'[^\[]'
# Use regex substitution to remove special characters from the 'text' column
probe_mapper["Forward_Sequence_formatted"] = probe_mapper["Forward_Sequence"].str.replace('[CG]', "CG")
```

```python
probe_mapper = probe_mapper.drop(columns=["length"])
probe_mapper
```

```python
def create_fasta(dataframe, sequence_column, identifier_column, output_file):
    with open(output_file, 'w') as f:
        for index, row in dataframe.iterrows():
            identifier = row[identifier_column]
            sequence = row[sequence_column]
            f.write(f'>{identifier}\n{sequence}\n')

df = probe_mapper #pd.DataFrame(data)

# Define columns
sequence_column = 'Forward_Sequence_formatted'
identifier_column = 'ID'

# Output file
output_file = 'output.fasta'

# Create FASTA file
create_fasta(df, sequence_column, identifier_column, output_file)
```
